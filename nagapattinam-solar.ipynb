Utility functions to abstract root path and file reading

```python
def get_rooted(stem):
    return "../input/lila-nagapattinam-dist/LiLa_Nagapattinam/" + stem

def read_df_UT(stem):
    return gpd.read_file(get_rooted(stem)).to_crs(epsg = 4326)
```
